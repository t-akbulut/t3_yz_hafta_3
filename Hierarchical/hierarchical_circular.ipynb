```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
```

```python

dataset = datasets.make_circles(n_samples=1000, factor=.5, noise=.05)[0]
```

```python
y = cluster.AgglomerativeClustering(n_clusters=4, affinity='euclidean', 
                                              linkage='ward').fit_predict(dataset)
```

```python
for idx,color in enumerate(['green','lightblue','orange','purple','teal']):
    plt.scatter(
        dataset[y ==idx,0], 
        dataset[y == idx,1], 
        s=20, 
        c=color
    )
```

```python
connect = kneighbors_graph(dataset, n_neighbors=5, include_self=False)
y = cluster.AgglomerativeClustering(n_clusters=2, affinity='euclidean', 
                                              linkage='complete',connectivity=connect).fit_predict(dataset)

for idx,color in enumerate(['green','lightblue','orange','purple','teal']):
    plt.scatter(
        dataset[y ==idx,0], 
        dataset[y == idx,1], 
        s=20, 
        c=color
    )
```
